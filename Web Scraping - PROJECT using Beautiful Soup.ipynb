# 1. Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#page 200 means scrapping is legal otherwise fraudelent.

page_wiki=requests.get("https://en.wikipedia.org/wiki/Main_Page")
page_wiki  
```

```python
page_wiki.content
```

```python
soup=BeautifulSoup(page_wiki.content, 'html.parser')
soup
```

```python
print(soup.prettify())
```

```python
header_h1 = soup.find_all('h1')
```

```python
header_h2 = soup.find_all('h2')
```

```python
header_h3 = soup.find_all('h3')
```

Let's extract texts from all headers

```python
#h1 texts
list_h1 = []
for i in header_h1:
    list_h1.append(i.text.replace('\n',''))
list_h1
```

```python
#h2 texts
list_h2 = []
for i in header_h2:
    list_h2.append(i.text.replace('\n',''))
list_h2
```

```python
#h3 texts
list_h3 = []
for i in header_h3:
    list_h3.append(i.text.replace('\n',''))
list_h3
```

```python
print("\n List_h1-",len(list_h1),"\n List_h2-",len(list_h2)," \n List_h3-",len(list_h3))
```

```python
import pandas as pd

wiki_h1=pd.DataFrame({})
wiki_h1['header-H1']=list_h1


wiki_h2=pd.DataFrame({})
wiki_h2['header-H2']=list_h2


wiki_h3=pd.DataFrame({})
wiki_h3['header-H3']=list_h3
```

```python
df=pd.concat([wiki_h1,wiki_h2,wiki_h3],axis=1)
df['header-H1'] = df['header-H1'].fillna(' ')
df['header-H2'] = df['header-H2'].fillna(' ')
df['header-H3'] = df['header-H3'].fillna(' ')
df
```
