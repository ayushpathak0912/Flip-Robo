# Bivariate analysis-

```python
for i in [x for x in datasheet.columns if x not in continous + ['1Gender of respondent']]:
    plt.figure(figsize=(15,4))
    sns.countplot(x=i,data=datasheet,hue='1Gender of respondent')
```

```python
tbl=pd.crosstab(datasheet['25 Convenient Payment methods'],datasheet['Website is as efficient as before'])
(tbl.div(tbl.sum(axis=1),axis=0)*100).plot(kind='bar',stacked=True,figsize=(16,4))
plt.xlabel('PaymentMethod',fontsize=12)
plt.ylabel('Efficiency',fontsize=12)
```

```python
tbl=pd.crosstab(datasheet['Security of customer financial information'],datasheet['Perceived Trustworthiness'])
(tbl.div(tbl.sum(axis=1),axis=0)*100).plot(kind='bar',stacked=True,figsize=(16,4))
plt.xlabel('Security of financial info',fontsize=12)
plt.ylabel('% of trust',fontsize=12)
```

```python
tbl=pd.crosstab(datasheet['25 Convenient Payment methods'],datasheet['Website is as efficient as before'])
(tbl.div(tbl.sum(axis=1),axis=0)*100).plot(kind='bar',stacked=True,figsize=(16,4))
plt.xlabel('PaymentMethod',fontsize=12)
plt.ylabel('Efficiency',fontsize=12)
```

```python
tbl=pd.crosstab(datasheet['Change in website/Application design'],datasheet['Frequent disruption when moving from one page to another'])
(tbl.div(tbl.sum(axis=1),axis=0)*100).plot(kind='bar',stacked=True,figsize=(16,4))
plt.xlabel('Web design',fontsize=12)
plt.ylabel('Frequent disruption',fontsize=12)
```

```python
tbl=pd.crosstab(datasheet['1Gender of respondent'],datasheet['Which of the Indian online retailer would you recommend to a friend?'])
(tbl.div(tbl.sum(axis=1),axis=0)*100).plot(kind='bar',stacked=True,figsize=(16,4))
plt.xlabel('Gender of the customer',fontsize=12)
plt.ylabel('Indian online retailer would you recommend to a friend',fontsize=12)
```
